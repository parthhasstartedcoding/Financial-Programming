```python
# prompt: print first n even numbers make function, give 2-3 methods

# Method 1: Using a loop and a counter
def print_even_numbers1(n):
  count = 0
  num = 2
  while count < n:
    print(num, end=" ")
    count += 1
    num += 2

# Method 2: Using a loop and a list
def print_even_numbers2(n):
  even_numbers = []
  for i in range(2, n * 2 + 1, 2):
    even_numbers.append(i)
  print(*even_numbers)  # Unpack the list for printing

# Method 3: Using a list comprehension
def print_even_numbers3(n):
  even_numbers = [i for i in range(2, n * 2 + 1, 2)]
  print(*even_numbers)  # Unpack the list for printing

# Example usage
n = int(input("Enter the number of even numbers to print: "))
print_even_numbers1(n)
print()
print_even_numbers2(n)
print()
print_even_numbers3(n)
```

```python
# prompt: Give me whole library details for strftime

import datetime

# List of all available format codes
format_codes = [
    "%a",  # Weekday short name (e.g., Mon)
    "%A",  # Weekday full name (e.g., Monday)
    "%w",  # Weekday as a number (0-6, Sunday is 0)
    "%d",  # Day of the month (01-31)
    "%b",  # Month short name (e.g., Jan)
    "%B",  # Month full name (e.g., January)
    "%m",  # Month as a number (01-12)
    "%y",  # Year short (e.g., 21)
    "%Y",  # Year full (e.g., 2021)
    "%H",  # Hour (24-hour clock) (00-23)
    "%I",  # Hour (12-hour clock) (01-12)
    "%p",  # AM/PM
    "%M",  # Minute (00-59)
    "%S",  # Second (00-59)
    "%f",  # Microsecond (000000-999999)
    "%z",  # UTC offset (e.g., +0100)
    "%Z",  # Time zone name (e.g., EST)
    "%j",  # Day of the year (001-366)
    "%U",  # Week number of the year (Sunday as the first day of the week) (00-53)
    "%W",  # Week number of the year (Monday as the first day of the week) (00-53)
    "%c",  # Local version of date and time
    "%x",  # Local version of date
    "%X",  # Local version of time
    "%%",  # A literal '%' character
]

# Print details for each format code
for code in format_codes:
    example = datetime.datetime.now().strftime(code)
    print(f"Format code: {code}, Example: {example}")
```
